
### Objective

We have 2 fields, our objective is to compare the yield between on different samples. (Using T value method) t - value = (Signal/Noise) = (Difference between the means/variability of group)

Formula = |X1-X2|/Sqrt((S1^2/n1)+(S2^2/n2))
T test is applied when

1) population Standard deviation is not available 2) n is less than 30



```python
import pandas as pd
import numpy as np
import math
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df=pd.read_excel(r"E:\Book1.xlsx")
df
```

```python
F1_mean = np.mean(df.d)
F2_mean = np.mean(df.e)

print(" Mean value of field 1 is {0} \n Mean value of field 2 is {1} ".format(F1_mean,F2_mean))

```

```python
F1_std = np.std(df.d)
F2_std = np.std(df.e)

print(" Standerd deviation of field 1 is {0} \n Standerd deviation of field 2 is {1} ".format(F1_std,F2_std))
```

```python
F1_var = F1_std*F1_std
F2_var = F2_std*F2_std
print(" Variance of field 1 is {0} \n Variance of field 2 is {1} ".format(F1_var,F2_var))
```

```python
n1=len(df.d)
n2=len(df.e)
```

```python
t_value = abs(F1_mean-F2_mean)/math.sqrt((F1_var/n1)+(F2_var/n2))
print("T - value is ",t_value)
```



H0 (Null hypothesis) - No Statistical difference between the samples. H1 (Alternate Hypothesis) - Exists a statistical difference betwen the samples.

If our t-value is less than critical -- fail to reject the null hypothesis. If our t-value is more than critical-- reject the null hypothesis.
finding the critical value

1) refer table for Df at rows and columns as significance (P value)

in our case we have 10 data points, 2 variable. So df is 20-2=18. Mostly used/ tested signifcance is .05. our case is two tailed and critical value is 2.101. Calculated T value is 1.36 which is lower  and accept the null hypothesis.

Conclusion "No statistical difference between the samples"
